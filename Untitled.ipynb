```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("car data.csv")
```

```python
data.head()
```

```python
#checking for null values
data.isnull().sum()
```

```python
data.describe()
```

```python
data['Year'] = 2022-data['Year']
```

```python
data.head()
```

```python
# handling categorical variables
final_data = data.drop(['Car_Name'],axis=1)
final_data = pd.get_dummies(final_data , drop_first = True)
```

```python
final_data.head()
```

```python
import seaborn as sns

sns.pairplot(final_data)
```

```python
#detecting outliers

import matplotlib.pyplot as plt

plt.scatter(final_data['Selling_Price'],final_data['Kms_Driven'])
plt.show()
```

```python
plt.scatter(final_data['Selling_Price'],final_data['Present_Price'])
plt.show()
```

```python
#removing outliers
final_data = final_data[final_data['Kms_Driven'] < 250000]
final_data = final_data[final_data['Present_Price'] < 50]
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_data['Selling_Price'],final_data['Kms_Driven'])
plt.show()
```

```python
plt.scatter(final_data['Selling_Price'],final_data['Present_Price'])
plt.show()
```

```python
# checking for correlated features
final_data.corr()
```

```python
sns.heatmap(final_data.corr() , cmap='RdYlGn')
```

```python
#dependent and independent features
y = final_data['Selling_Price']
x = final_data.drop(['Selling_Price'] , axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
#feature selection using ANOVA f-value  
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif


fvalue = SelectKBest(f_classif, k = 'all')
x_best = fvalue.fit_transform(x,y)
```

```python
for i in range(0,len(x.columns)):
    print('{} : {}'.format(x.columns[i],fvalue.scores_[i]))
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#training using Random Forest using Randomized Search CV

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model = rf.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics

mse = metrics.mean_squared_error(y_test,y_pred)
mae = metrics.mean_absolute_error(y_test,y_pred)
```

```python
print('mse:',mse)
print('mae:',mae)
```

```python
#training using Decison Tree

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model2 = dt.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
mse2 = metrics.mean_squared_error(y_test,y_pred2)
mae2 = metrics.mean_absolute_error(y_test,y_pred2)
```

```python
print('mse:',mse2)
print('mae:',mae2)
```

```python
#training using linear Regression

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model3 = lr.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
y_pred3 = model3.predict(x_test)
```

```python
mse3 = metrics.mean_squared_error(y_test,y_pred3)
mae3 = metrics.mean_absolute_error(y_test,y_pred3)
```

```python
print('mse:',mse3)
print('mae:',mae3)
```
